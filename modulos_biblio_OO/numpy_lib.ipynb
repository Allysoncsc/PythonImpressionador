```python
import matplotlib.pyplot as plt
import numpy as np

#                           min,max,qtd
vendas = np.random.randint(100,300,50)
meses = np.arange(1,51)

plt.axis([0,50,0,max(vendas)+50])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.plot(meses,vendas, color = 'green')

plt.show()
```

```python
plt.figure(1, figsize=(15, 3))
plt.subplot(1, 3, 1)
plt.plot(meses, vendas, color='red')
plt.subplot(1, 3, 2)
plt.scatter(meses, vendas)
plt.subplot(1, 3, 3)
plt.bar(meses, vendas)
plt.show()
```
